#  Cleaning the Data

```python
#Deleting useless columns
train = train.drop(columns=['PassengerId','Name','Cabin','Ticket'], axis = 1)
test = test.drop(columns=['PassengerId','Name', 'Cabin','Ticket'], axis = 1)
```

```python
train['Age'] = pd.to_numeric(train['Age'], errors='coerce')
```

```python
# fill  na data
train.fillna(train.Age.median(), inplace=True)
test.fillna(train.Age.median(), inplace=True)
```

```python
# some fature engineer  to build a more meaniful feature
# With this we can analyse the influence of having family of who has in titanic
train['hasFamily'] = ((train.Parch + train.SibSp)>0)*1
test['hasFamily'] = (test.Parch + train.SibSp > 0) *1
```

```python
train = train.drop(columns=['Parch', 'SibSp'], axis = 1)
test = test.drop(columns=['Parch', 'SibSp'], axis = 1)
```

```python
def filter_age(x):
    if x < 1:
        return 0
    else:
         return x
```

```python
train['Age'] = train['Age'].apply(lambda x:  filter_age(x))
```

```python
# before training the model we need  encode non-numerical fetures
from sklearn.preprocessing import OneHotEncoder

sexDummies_train = pd.get_dummies(train.Sex, prefix='Sex')
embarkedDummies_train = pd.get_dummies(train.Embarked, prefix='Embarked')
sexDummies_test = pd.get_dummies(test.Sex, prefix='Sex')
embarkedDummies_test = pd.get_dummies(test.Embarked, prefix='Embarked')


train = pd.concat([train,sexDummies_train,embarkedDummies_train], axis=1)
test = pd.concat([test,sexDummies_test,embarkedDummies_test], axis=1)
```

```python
train
```

```python
train.drop(columns='Embarked_28.0', inplace=True)
```

```python
# We don't need  Embarked and Sex columns anymore, let's get rid of them
train.drop(columns=['Sex','Embarked'], inplace=True)
test.drop(columns=['Sex','Embarked'], inplace = True)
```

# Training and evaluating models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier()
```

```python
X_train = train.drop(columns='Survived')
y_train = train['Survived']
X_test = test.drop(columns='Survived')
y_test= test['Survived']
```

```python

dt_clf.fit(X_train, y_train)
rf_clf.fit(X_train,y_train)
```

```python
ypred_dt = dt_clf.predict(X_test)
ypred_rf = rf_clf.predict(X_test)
```

```python
# lets evaluating both models
from sklearn.metrics import roc_curve, roc_auc_score, \
confusion_matrix, precision_recall_curve, precision_score, recall_score, f1_score
```

```python
cm_dt = confusion_matrix(y_test,ypred_dt)
cm_rf = confusion_matrix(y_test, ypred_rf)
```

```python
pd.DataFrame(cm_dt, columns=['Survived','Not Survived'], index=['Survived','Not Survived'])
```

```python
pd.DataFrame(cm_rf, columns=['Survived','Not Survived'], index=['Survived','Not Survived'])
```

```python
print("Decision Tree Recall/Precision metrics")
print("Recall: %.2f"%recall_score(y_test, ypred_dt))
print("Precision: %.2f"%precision_score(y_test, ypred_dt))
```

```python
print("Random Forest Recall/Precision metrics")
print("Recall: %.2f"%recall_score(y_test, ypred_rf))
print("Precision: %.2f"%precision_score(y_test, ypred_rf))
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
yscore_dt = cross_val_predict(dt_clf ,X_train, y_train,cv=5,method="predict_proba")[:,1]
yscore_rf = cross_val_predict(rf_clf ,X_train, y_train,cv=5, method="predict_proba")[:,1]

```

```python
fpr_dt, tpr_dt, thresholds = roc_curve(y_train, yscore_dt)
fpr_rf, tpr_rf, thresholds = roc_curve(y_train, yscore_rf)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1],[0,1],'k--')
```

```python
plot_roc_curve(fpr_rf,tpr_rf,"Random Forest")
plt.title("ROC Decision Tree")
plt.plot(fpr_dt,tpr_dt, "b:", label="Decision Tree")
plt.show()
```

```python
# let's comparing the roc_scores 
print("Decision tree ROC score: %.2f"%roc_auc_score(y_train,yscore_dt))
print("Random Forest ROC score: %.2f"%roc_auc_score(y_train,yscore_rf))
```

#Tunneling the model

```python
from sklearn.model_selection import GridSearchCV
```

```python

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
rf_search = GridSearchCV(RandomForestClassifier(),random_grid,cv=5, verbose=2)
```

```python
rf_search.fit(X_train,y_train)
```

```python
rf_search.best_estimator_
```

```python
final_model =RandomForestClassifier(bootstrap=False, max_depth=10, max_features='sqrt',
                       min_samples_leaf=4, n_estimators=400)
final_model.fit(X_train, y_train)
```

```python
ypred = final_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, ypred)
pd.DataFrame(cm, columns=['Survived','Not Survived'], index=['Survived','Not Survived'])
```
